```python
import matplotlib.pyplot as plt


epochs = []
losses = []
mses = []
maes = []


with open("electricity_log.txt", "r") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        try:
            parts = line.split(":")
            epoch_part = parts[0]
            epoch_str = epoch_part.split()[1].split("/")[0] 
            epoch = int(epoch_str)
            epochs.append(epoch)

            stats_part = parts[1]
            stat_items = stats_part.split(",")
            loss_value = float(stat_items[0].split("=")[1].strip())
            mse_value = float(stat_items[1].split("=")[1].strip())
            mae_value = float(stat_items[2].split("=")[1].strip())
            losses.append(loss_value)
            mses.append(mse_value)
            maes.append(mae_value)
        except Exception as e:
            print("解析错误:", line, e)

# 绘制 Loss vs Epoch
plt.figure()
plt.plot(epochs, losses, marker='o', color='b')
plt.title("Loss vs Epoch")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.savefig("loss_vs_epoch.png")
plt.show()

# 绘制 MSE vs Epoch
plt.figure()
plt.plot(epochs, mses, marker='o', color='r')
plt.title("MSE vs Epoch")
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.grid(True)
plt.savefig("mse_vs_epoch.png")
plt.show()

# 绘制 MAE vs Epoch
plt.figure()
plt.plot(epochs, maes, marker='o', color='g')
plt.title("MAE vs Epoch")
plt.xlabel("Epoch")
plt.ylabel("MAE")
plt.grid(True)
plt.savefig("mae_vs_epoch.png")
plt.show()
```
